## Rate of Crystal Growth
$$v_{ls} = v_0 e^\frac{-\Delta G^\text{act diff}}{kT}$$
$$v_{sl} = v_0 e^\frac{-(\Delta G_f +\Delta G^\text{act diff}}{kT} = v_{ls}e^\frac{-\Delta G_f}{kT}$$

The net rate is the difference between the above, so...

$$u = a_0 f (v_{ls} - v_{sl}) = a_0 f v_0  e^\frac{-\Delta G^\text{act diff}}{kT} \left[1- e^\frac{-\Delta G_f}{kT} \right]$$

$$D = a^2_0 v_{ls} = \frac{kT}{2\pi a_0 \eta}$$

so,

$$v_{ls} = v_0 e^\frac{-\Delta G^\text{act diff}}{kT} = \frac{D}{a_0^2}$$

and

$$u = a_0 f (v_{ls} - v_{sl}) = a_0 f v_0  e^\frac{-\Delta G^\text{act diff}}{kT} \left[1- e^\frac{-\Delta G_f}{kT} \right] = \frac{fkT}{2\pi a^2_0 \eta} \left[1- e^\frac{-\Delta G_f}{kT} \right]$$



## Rate of Nucleation

## Rough and Smooth Interfaces

$$\Delta G_f = (T_m - T)\Delta S_F  < k T_m $$

$$u = \frac{u\Delta G_f}{2\pi a^2_0 \eta}= \frac{u\Delta S_f}{2\pi a^2_0 \eta}(T_m-T $$

$$\Delta G = \Delta H_a + \Delta H_L - T\Delta S_a - T\Delta S_L$$

Adsorbiton

$$\Delta H_a = -N_a \Delta H_f\frac{2 \eta_0}{B}$$

Lateral
$$\Delta H_L = -N_a \Delta H_f\frac{\eta_L}{B}\frac{N_a}{N}$$

Entropy
$$\Delta S_a = - \frac{\Delta H_f N_a}{T_m} $$

$$\Delta S_L = kN\log\left(\frac{N}{N-N_a}\right)+kN_a\log\left( \frac{N-N_a}{N}\right)$$

$$\Delta G = -N_a^2 \Delta H_f\frac{2 \eta_0}{BN} +N_a \Delta H_f \left(\frac{T}{T_m}+\frac{\eta_L}{B}\right)+kTN_a\log\left(\frac{N_a}{N-N_a}\right)-kTN\log\left( \frac{N}{N-N_a}\right)$$

$$\Delta G = (1-x)x\alpha - \frac{x(1-\theta)\Delta H_f}{kT_m}+(1-x)\theta\log(1-x)+x\theta\log(x)$$

$$\alpha = \frac{\Delta H_f}{kT_m}\frac{\eta_L}{B}$$



```python
x = linspace(0,1,100)
figure(figsize=(15,6))
for alpha in linspace(1,5,5):
    dGnorm = (1-x)*x*alpha+(1-x)*log(1-x)+x*log(x)
    plot(x,dGnorm,'k')
    annotate('$\\frac{\Delta S}{R} = %.1f$' % alpha,xy=(.5,dGnorm[50]+.01),ha='center')

xlabel("Sites Occupied")
ylabel("Free Energy Change")
show()

```

## Kinetics of a Phase Transformation

$$(V-V_C)I_V(t')dt' $$
$$\frac{4}{3}\pi u^3(t-t')^3$$

$$dV_C = \frac{4}{3}\pi u^3(t-t')^3 (V-V_c)I_V(t')dt' $$

V_c(t) =\int_0^t \frac{4}{3}\pi u^3(t-t')^3  (V-Vc)I_V(t')dt'

Assume constant Growth

$$\frac{V_C}{V} = 1-e^{-\frac{\pi}{3}I_vu^3t^4} $$


==> only for the isothermal case!


```python
t = linspace(0,1,100)
u = 2 #arb
Iv = 2 #arb
figure(figsize=(15,6))

vcbyv = 1-exp(-(pi/3)*Iv*u**3*t**4)
plot(t,vcbyv,'k')
xlabel("Time (au)")
ylabel("Transformation Extent")
```
